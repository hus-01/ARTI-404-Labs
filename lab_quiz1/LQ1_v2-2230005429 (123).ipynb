```python
#Add or remove libraries as needed
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

### Task#1: Load/read cat.jpeg as a grayscale image [1 mark]

```python
#load the image as a grayscale
cat_img = cv2.imread('cat.jpeg', cv2.IMREAD_GRAYSCALE)

#display image
plt.imshow(cat_img, cmap='gray')
plt.axis('off')
plt.show()
```

### Task#2: Apply thresholding (thresh=100) to the image and display the results [1.5 mark]
#### Note: use cmap = 'gray' to properly display the thresholded image

```python
#apply threshold = 100
ret, thresh_img = cv2.threshold(cat_img, 100, 255, cv2.THRESH_BINARY)

#display image
plt.figure()

plt.subplot(1,2,1)
plt.imshow(cat_img, cmap='gray')
plt.title('original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(thresh_img, cmap='gray')
plt.title('thresholded = 100')
plt.axis('off')

plt.show()
```


### Task#3: Load/read the camera man, blur it using a 9x9 gaussian filter (sigma=5), and display the results.   [2.5 marks]

```python
#load image in grayscale
cam_img = cv2.imread('cameraman.tif', cv2.IMREAD_GRAYSCALE)

#apply Gaussian filter
blurred_img = cv2.GaussianBlur(cam_img, (9, 9), sigmaX=5 , sigmaY=5)

#display results
plt.figure()

plt.subplot(1,2,1)
plt.imshow(cam_img, cmap='gray')
plt.title('original Cameraman')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(blurred_img, cmap='gray')
plt.title('Gaussian Blur (9x9 and sigma = 5)')
plt.axis('off')

plt.show()
```
