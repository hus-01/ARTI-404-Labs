```python
#SubTask 1: Unsharp Masking
# Read the image
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('images/cameraman.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

#gussian sigma
sigma=2

#parameter
amount=1.5

# Convert image to float32 for processing
image_float = image.astype(np.float32) / 255.0

# Apply Gaussian blur
blurred = cv2.GaussianBlur(image_float, (0, 0), sigmaX=sigma, sigmaY=sigma)

# Perform Unsharp Masking
sharpened = np.clip(image_float + amount * (image_float - blurred), 0, 1)
# Plot original, blurred, and sharpened images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(blurred)
ax[1].set_title("Blurred Image")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened (Unsharp Masking)")
ax[2].axis("off")

plt.show() 
```

```python
#Task 1
# Read the image
image = cv2.imread('images/pic1.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Convert image to float32 for processing
image_float = image.astype(np.float32) / 255.0

# Apply 7x7 Box Filter (Averaging filter)
smoothed = cv2.blur(image_float, (7, 7))

# Plot images side by side
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(smoothed)
ax[1].set_title("Smoothed Image (7x7 Box Filter)")
ax[1].axis("off")

plt.show()
```

```python
#Task 2
# Read the image
image = cv2.imread('images/pic.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert image to float32
image_float = image.astype(np.float32) / 255.0

# Apply 5x5 Gaussian filter
gaussian_5 = cv2.GaussianBlur(image_float, (5, 5), 0)

# Apply 21x21 Gaussian filter
gaussian_21 = cv2.GaussianBlur(image_float, (21, 21), 0)

# Plot images side by side
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gaussian_5)
ax[1].set_title("Gaussian Filter (5x5)")
ax[1].axis("off")

ax[2].imshow(gaussian_21)
ax[2].set_title("Gaussian Filter (21x21)")
ax[2].axis("off")

plt.show()
```

```python
#Task 3
# Read the image
image = cv2.imread('images/pic1.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert image to float32
image_float = image.astype(np.float32) / 255.0

# Apply Laplacian filter
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

# Sharpen the image
sharpened = np.clip(image_float - laplacian, 0, 1)

# Plot images side by side
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(np.clip(laplacian, 0, 1))
ax[1].set_title("Laplacian Image")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened Image (Laplacian)")
ax[2].axis("off")

plt.show()
```
