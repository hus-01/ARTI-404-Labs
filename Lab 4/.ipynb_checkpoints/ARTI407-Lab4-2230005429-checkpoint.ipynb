```python
import cv2

# Load the image in grayscale
img = cv2.imread('images/cameraman.png', cv2.IMREAD_GRAYSCALE)

# Set the threshold values
threshold_values = [0, 50, 100, 150, 200]

# Apply thresholding with different threshold values
for threshold_value in threshold_values:
    ret, thresh = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)
    cv2.imshow(f'Thresholded Image ({threshold_value})', thresh)

# Display the original image
cv2.imshow('Original Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, img_as_float
from skimage import exposure

img = data.moon()

# Contrast stretching
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))

fig = plt.figure(figsize=(10, 7)) 
fig.add_subplot(2, 1, 1) 
plt.imshow(img, cmap = 'gray') 
plt.axis('off') 
plt.title('Original image') 
fig.add_subplot(2, 1, 2) 
plt.hist(img.flat, bins = 256, range=(0, 255)) 
plt.xlabel('gray level values') 
plt.ylabel('Number of pixels') 
plt.show() 



fig = plt.figure(figsize=(10, 7)) 	
fig.add_subplot(2, 1, 1) 
plt.imshow(img_rescale, cmap = 'gray') 
plt.axis('off') 
plt.title('Contrast stretch')
fig.add_subplot(2, 1, 2) 
plt.hist(img_rescale.flat, bins = 256, range=(0, 255)) 
plt.xlabel('gray level values') 
plt.ylabel('Number of pixels') 
plt.show()
```

```python
#Task 1

import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure

# Load moon image
img = data.moon()

# 3rd and 80th percentiles
p3, p80 = np.percentile(img, (3, 80))

# Rescale intensity (contrast stretching using chosen percentiles)
img_rescale_3_80 = exposure.rescale_intensity(img, in_range=(p3, p80))

# Display image + histogram
plt.figure(figsize=(10, 4))
plt.imshow(img_rescale_3_80, cmap='gray')
plt.axis('off')
plt.title('Moon: Rescaled intensity (3rd80th percentiles)')
plt.show()

plt.figure(figsize=(10, 4))
plt.hist(img_rescale_3_80.ravel(), bins=256, range=(0, 255))
plt.title('Histogram after rescaling (3rd80th percentiles)')
plt.xlabel('Gray level')
plt.ylabel('Number of pixels')
plt.show()
```

```python
#Task 2

img = data.moon()

# Histogram equalization
img_eq = exposure.equalize_hist(img)

# Display image
plt.figure(figsize=(10, 4))
plt.imshow(img_eq, cmap='gray')
plt.axis('off')
plt.title('Moon: Histogram Equalization (equalize_hist)')
plt.show()

# Display histogram (range 0..1 because output is float)
plt.figure(figsize=(10, 4))
plt.hist(img_eq.ravel(), bins=256, range=(0, 1))
plt.title('Histogram after equalize_hist (flattened)')
plt.xlabel('Intensity (0 to 1)')
plt.ylabel('Number of pixels')
plt.show()
```

```python
#Task 3

# Source and reference images
source = data.chelsea()   # image to transform
reference = data.rocket() # reference look

# Histogram matching
matched = exposure.match_histograms(source, reference, channel_axis=-1)

# Display results
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(source)
plt.axis('off')
plt.title('Source: chelsea')

plt.subplot(1, 3, 2)
plt.imshow(reference)
plt.axis('off')
plt.title('Reference: rocket')

plt.subplot(1, 3, 3)
plt.imshow(matched.astype(np.uint8))
plt.axis('off')
plt.title('Matched Result')

plt.tight_layout()
plt.show()
```
