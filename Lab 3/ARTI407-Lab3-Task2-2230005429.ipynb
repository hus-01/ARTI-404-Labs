```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


#load the image

image = cv2.imread("images/lena_gray.png", cv2.IMREAD_GRAYSCALE)
```

```python
#Task 1: negative transformation

#subtract pixel values from 255
Negative_image = 255- image

plt.figure(figsize=(10,5))

#show the image before and after using Matplotlib
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(Negative_image, cmap='gray')
plt.title("Negative")
plt.axis('off')

plt.show()
```

```python
#Task 2: log transformation

# scaling constant
c = 1.0
#apply log transform
Log_image = c * np.log1p(image)

#show the image before and after using Matplotlib
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(Log_image, cmap='gray')
plt.title("Log")
plt.axis('off')

plt.show()
```

```python
#Task 3: power law transformation

#Gamma < 1: Brightens image
#Gamma > 1: Darkens image

#brighten (gamma = 0.3)
gamma_image_Bright = np.power(image / 255.0, 0.3) * 255
#darken (gamma = 4.0)
gamma_image_Dark = np.power(image / 255.0, 4.0) * 255

#show the image before and after using Matplotlib
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gamma_image_Bright, cmap='gray')
plt.title("Power law - Bright")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gamma_image_Dark, cmap='gray')
plt.title("Power law - Dark ")
plt.axis('off')

plt.show()
```
