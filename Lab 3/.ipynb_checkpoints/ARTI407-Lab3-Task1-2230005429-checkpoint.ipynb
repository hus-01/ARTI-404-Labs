```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def scale_up_image(image, scale_factor):
    """
    Increases the size of the image by the given scale factor.
    Args:
        image (numpy array): Original image.
        scale_factor (float): Scaling factor.
    Returns:
        numpy array: Scaled image.
    """
    height, width = image.shape[:2]
    scaled_image = cv2.resize(
        image,
        (int(width * scale_factor), int(height * scale_factor)),
        interpolation=cv2.INTER_LINEAR
    )
    return scaled_image


def rotate_image(image, angle):
    """
    Rotates the image by the given angle using getRotationMatrix2D.
    Args:
        image (numpy array): Original image.
        angle (float): Rotation angle in degrees.
    Returns:
        numpy array: Rotated image.
    """
    height, width = image.shape[:2]
    center = (width // 2, height // 2)

    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated_image = cv2.warpAffine(image, M, (width, height))

    return rotated_image


def shear_image(image, shear_factor):
    """
    Performs shear transformation on the image.
    Args:
        image (numpy array): Original image.
        shear_factor (float): Shear intensity.
    Returns:
        numpy array: Sheared image.
    """
    height, width = image.shape[:2]

    M = np.float32([
        [1, shear_factor, 0],
        [0, 1, 0]
    ])

    sheared_image = cv2.warpAffine(image, M, (width + int(height * shear_factor), height))

    return sheared_image


def plot_images(original, scaled, rotated, sheared):
    """
    Displays images side by side.
    """
    plt.figure(figsize=(14, 4))

    plt.subplot(1, 4, 1)
    plt.imshow(original, cmap='gray')
    plt.title("Original")
    plt.axis('off')

    plt.subplot(1, 4, 2)
    plt.imshow(scaled, cmap='gray')
    plt.title("Scaled")
    plt.axis('off')

    plt.subplot(1, 4, 3)
    plt.imshow(rotated, cmap='gray')
    plt.title("Rotated 120")
    plt.axis('off')

    plt.subplot(1, 4, 4)
    plt.imshow(sheared, cmap='gray')
    plt.title("Sheared")
    plt.axis('off')

    plt.show()


# ---------------- Main ----------------

image_path = "images/lena_gray.png"

original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if original_image is None:
    print("Error loading image.")

scaled_image = scale_up_image(original_image, 1.5)
rotated_image = rotate_image(original_image, 120)
sheared_image = shear_image(original_image, 0.4)

plot_images(original_image, scaled_image, rotated_image, sheared_image)
```
