```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the image

image = cv2.imread("images/lena_gray.png", cv2.IMREAD_GRAYSCALE)

```

```python
#Task 1: Increase the size of the image

resized_image = cv2.resize(image, (400, 400))
cv2.imshow("lena_gray1", image)
cv2.imshow("lena_gray2", resized_image)
cv2.waitKey()
```

```python
#Task 2: Applay 120 degrees rotation

#get the image dimensions
height, width = image.shape[:2]
 
#find the rotation center
center = (width // 2, height // 2)

#get the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1.0)
 
#perform rotation
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

#show the image before and after using Matplotlib
plt.figure(figsize=(6,6))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(rotated_image, cmap='gray')
plt.title("Rotated 120")
plt.axis('off')

plt.show()
```

```python
#Task 3: perform Shear operations over the image

image = cv2.imread("images/lena_gray.png", cv2.IMREAD_GRAYSCALE)
height, width = image.shape[:2]

#shear factor
shear_factor = 0.5

#shear matrix (horizontal shear)
M = np.float32([[1, shear_factor, 0],
                [0, 1, 0]])

#apply shear transformation
sheared_img = cv2.warpAffine(image, M, (int(width + shear_factor*height), height))

#show the image before and after using Matplotlib
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sheared_img, cmap='gray')
plt.title("Sheared Image")
plt.axis('off')

plt.show()
```
